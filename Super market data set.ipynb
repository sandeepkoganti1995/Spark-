```
import pyspark
from pyspark.sql import SparkSession
```

```
new_Session=SparkSession.builder.appName("supermarket").getOrCreate()
```

```
df=spark.read.csv("/FileStore/tables/supermarket_sales___Sheet1.csv", header=True,inferSchema=True)
```

```
df.show(5)
```

```
df.select(df.columns[:5]).show(5)
```

```
df=df.drop("Invoice ID")
```

```
df.show()
```

```
from pyspark.sql.functions import countDistinct
df.select(countDistinct("Branch")).show()
```

```
df.groupBy("Branch").count().show() # to get count in the column
```

```
df.groupBy("City").count().show()
```

```
df.groupBy("Customer type").count().show()
```

```
df.groupBy("Product line").count().show() # they are selling health and beauty products least
```

```
df.groupBy("Payment").count().show()
```

```
df.groupBy("Gender").count().show() # almost both the genders are shopping at equal rate
```

```
df=df.drop("Date","time")
```

```
df.show()
```

```
df1=df.groupBy("Product line").sum()
df1.select("Product line","sum(gross income)","sum(Quantity)").show() # sales are higher in food and bevarages
```

```
df2=df.groupBy("Product line","Payment").count().show()
```

```
df.groupBy("Payment").sum().show() # the number of people uing the credit card is low when compared to other payment methods
```
