```
import pyspark
from pyspark.sql import SparkSession
new_session=SparkSession.builder.appName("Joins").getOrCreate()
```

```
from pyspark.sql.types import StructType, StructField,StringType,IntegerType
empnames=[("James","","Smith","36636","M",3000),
    ("Michael","Rose","","40288","M",4000),
    ("Robert","","Williams","42114","M",4000),
    ("Maria","Anne","Jones","39192","F",4000),
    ("Jen","Mary","Brown","","F",-1)
  ]

schema=StructType([ \
    StructField("firstname",StringType(),True), \
    StructField("middlename",StringType(),True), \
    StructField("lastname",StringType(),True), \
    StructField("id", StringType(), True), \
    StructField("gender", StringType(), True), \
    StructField("salary", IntegerType(), True) \
  ])

df=spark.createDataFrame(data=empnames,schema=schema)
```

```
df.show()
```

```
dept = [("Finance",10), \
    ("Marketing",20), \
    ("Sales",30), \
    ("IT",40) \
  ]
deptColumns = ["dept_name","dept_id"]
deptDF = spark.createDataFrame(data=dept, schema = deptColumns)
deptDF.printSchema()
deptDF.show(truncate=False)
```

```
empDF.join(deptDF,empDF.emp_dept_id==deptDF.dept_id,"inner").show() # inner joins combines the columns when there is a match
```

```
empDF.join(deptDF,empDF.emp_dept_id ==  deptDF.dept_id,"inner") \
     .show(truncate=False)
```

```
emp = [(1,"Smith",-1,"2018","10","M",3000), \
    (2,"Rose",1,"2010","20","M",4000), \
    (3,"Williams",1,"2010","10","M",1000), \
    (4,"Jones",2,"2005","10","F",2000), \
    (5,"Brown",2,"2010","40","",-1), \
      (6,"Brown",2,"2010","50","",-1) \
  ]
empColumns = ["emp_id","name","superior_emp_id","year_joined", \
       "emp_dept_id","gender","salary"]

empDF = spark.createDataFrame(data=emp, schema = empColumns)
empDF.printSchema()
empDF.show(truncate=False)
```

```
empDF.join(deptDF,empDF.emp_dept_id ==  deptDF.dept_id,"left").show()
```

```
empDF.show()
deptDF.show()
```

```
empDF.join(deptDF,empDF.emp_dept_id ==  deptDF.dept_id,"right").show()
```

```
empDF.join(deptDF,empDF.emp_dept_id ==  deptDF.dept_id,"leftsemi").show() 
```

```
empDF.show()
deptDF.show()
```

```
empDF.join(deptDF,empDF.emp_dept_id ==  deptDF.dept_id,"leftanti").show() 
```

```
# joins using sql

empDF.createOrReplaceTempView("emp")
deptDF.createOrReplaceTempView("dept")
```

```
spark.sql('SELECT * from emp e,dept d where e.emp_dept_id == d.dept_id').show()
```

```
spark.sql("select * from EMP e RIGHT JOIN DEPT d ON e.emp_dept_id == d.dept_id") \
  .show(truncate=False)
```
